# Notebook to showcase correlation and randomness

This notebook showcases how random scores when applying the knapsack post-processing can also lead to state of the art scores in the SumMe dataset

```python
import h5py
import numpy as np
import json 
from Utils import generate_summary_single,knapsack_wrapper_with_rating,correlation_single_pred
```

```python
splits = json.load(open(('Splits/summe_can_1.json')))

```

Testing the Performance on a Single Video

```python
summe_dataset = h5py.File('Data/googlenet/googlenet_summe.h5')

avg_corr = []

avg_corr_post = []
summe_prediction_size = summe_dataset['video_8']['gtscore'][...].shape
for i in range(200):

    prediction = np.abs(np.random.normal(0,1,size=summe_prediction_size))
    avg_corr.append(correlation_single_pred(prediction,'video_8',summe_dataset,'summe')[0])
    avg_corr_post.append(knapsack_wrapper_with_rating(prediction,'video_8',summe_dataset,'summe')[0])


print(np.max(avg_corr))
print(np.mean(avg_corr))

print(np.max(avg_corr_post))
print(np.mean(avg_corr_post))
```

Test over the whole SumMe dataset

```python
overall_split_performance =[]
overall_split_performance_best = []
overall_split_performance_post =[]
overall_split_performance_post_best = []
for split in splits:
    split_performance =[]
    split_performance_best = []
    split_performance_post = []
    split_performance_post_best = []

    split = split['test_keys']

    for test_index in split:
        avg_corr = []
        avg_corr_post = []
        print(test_index.split('/')[1])
        summe_prediction_size = summe_dataset[test_index.split('/')[1]]['gtscore'][...].shape
        for i in range(200):

            prediction = np.clip(np.abs(np.random.normal(0,1,size=summe_prediction_size)),0,1)
            avg_corr.append(correlation_single_pred(prediction,test_index.split('/')[1],summe_dataset,'summe')[0])
            avg_corr_post.append(knapsack_wrapper_with_rating(prediction,test_index.split('/')[1],summe_dataset,'summe')[0])
        print(np.max(avg_corr))
        print(np.mean(avg_corr))

        print(np.max(avg_corr_post))
        print(np.mean(avg_corr_post))
        split_performance.append(np.mean(avg_corr))
        split_performance_best.append(np.max(avg_corr))
        split_performance_post.append(np.mean(avg_corr_post))
        split_performance_post_best.append(np.max(avg_corr_post))

    overall_split_performance.append(split_performance)
    overall_split_performance_best.append(split_performance_best)
    overall_split_performance_post.append(split_performance_post)
    overall_split_performance_post_best.append(split_performance_post_best)


print(f'Performance of average random without Knapsack post processing: {np.mean(overall_split_performance)} ')
print(f'Performance of average random with best pick without Knapsack post processing: {np.mean(overall_split_performance_best)} ')
print(f'Performance of average random  with Knapsack post processing: {np.mean(overall_split_performance_post)}')
print(f'Performance of average random with best pick with Knapsack post processing: {np.mean(overall_split_performance_post_best)}')



        
```
