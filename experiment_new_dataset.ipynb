```python
import numpy as np
from sklearn.decomposition import PCA
import json
from Utils import kts,calculate_metrics
from numpy import linalg as LA
import h5py
import time
import json
```

```python
autoshot_results  = {}
autoshot_ext = "autoshot_part_"
```

```python
for i in range(1,8):
    features = h5py.File(f'{autoshot_ext}{i}.hdf5')
    for key in list(features.keys()):
        feature = features[key]
        feature = [feat/LA.norm(feat) for feat in feature]
        n_frames = len(feature)
        start = time.time()
        print(f'Start KTS for sequence length {n_frames}')
        change_points,_ ,_= kts(n_frames,np.array(feature),vmax=1.0)
        autoshot_results[key] = change_points.tolist()
        print(f'Time taken for KTS is : {time.time()-start}')

json.dump(autoshot_results,open('autoshot_results.json','w'))
```
